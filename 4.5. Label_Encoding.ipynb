# Label Encoding:
-    converting the labels into numeric form.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

## Label encoding of Breast Cancer Dataset

```python
# loading the data from csv file to pandas data frame
cancer_data = pd.read_csv("/content/breast_cancer_data.csv")
```

```python
cancer_data.head()
```

```python
# finding the count of different labels
cancer_data['diagnosis'].value_counts()
# or
cancer_data.diagnosis.value_counts()
```

```python
# load the Label Encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the Data Frame
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

---

**0 -> Benign**

**1 -> Malignant**

---

Arranged in alphabetical order and assigns numbers.

```python
cancer_data.target.value_counts()
```

## Label encoding of Iris Dataset

```python
iris_data = pd.read_csv("/content/iris_data.csv")
```

```python
iris_data.head()
```

```python
iris_data.Species.value_counts()
# or
iris_data['Species'].value_counts()
```

```python
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(iris_data.Species)
```

```python
iris_data['target'] = labels
```

```python
iris_data.head()
```

---

**0 -> Iris-setosa**
**1 -> Iris-versicolor**
**2 -> Iris-virginica**

---

```python
# always use dictionary type syntax
iris_data['target'].value_counts()
```

```python
# this type of syntax is not reliable , avoid using this.
iris_data.target.value_counts()
```
