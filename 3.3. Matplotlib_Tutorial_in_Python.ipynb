# Matplotlib:
-    Useful for making Plots

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

## Plotting the Data

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angel')
plt.ylabel('sin value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g--')
plt.show()
```

```python
plt.plot(x,y,'b*')
plt.show()
```

```python
# Multiple graphs
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

## Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'JAVA', 'PYTHON', 'GO']
students = [23,34,56,65,15]
ax.bar(langs, students, color='skyblue', edgecolor='black', width=0.6)
plt.xlabel('Lamguages')
plt.ylabel('number of Students')
plt.show()
```

## Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'JAVA', 'PYTHON', 'GO']
students = [23,34,56,65,15]
ax.pie(students, labels=langs, autopct='%1.1f%%')
plt.show()
```

## Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

## 3D Scatter Plot

```python
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
