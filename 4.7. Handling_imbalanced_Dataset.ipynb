# Imbalanced Dataset:
-    A dataset with an unequal class distribution.

```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is HIGHLY IMBALANCED DATASET.

---

 **0 -> Legit Transactions**

 **1 -> Fraudulent Transactions**

---

```python
# seperating the legit and fraud transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

### Under-Sampling

Build a sample dataset containing similar distribution of Legit and Fraudulent transactions.

Number of Fraudulent Transactions -> 492

New Number of Legit Transactions -> 492 (around 500)

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the both data frames

```python
new_credit_card_data = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_credit_card_data.head()
```

```python
new_credit_card_data.tail()
```

```python
new_credit_card_data['Class'].value_counts()
```
